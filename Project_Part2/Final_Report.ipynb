# Methodology

&#9989; TensorFlow Model Optimization is a toolkit for optimizing machine learning models to improve their performance, reduce their size, and make them more efficient to deploy on a variety of platforms. It provides a range of techniques for optimizing models, including pruning, quantization, and compression. The method I am going to use is weight pruning. Pruning of a model involves the removal of parameters within the models that have minimal impact on its accuracy of predictions which can reduce the size of the model and improve its performance.This is done by setting individual parameters to zero and making the network sparse. This would lower the number of parameters in the model while keeping the architecture the same. pruning can reduce the number of parameters and operations in a model, leading to faster inference times and lower memory usage.

The time of running it on GPU decreased a little bit after weight pruning but not significantly.
